2.Are there any abnormal products against rule:1 product_id is only owed by 1 team


```python
def check_abnormal_products(df):
    abnormal_products = []
    unique_products = df['product_id'].unique()

    for i in unique_products:
        teams = df[df['product_id'] == i]['team_own'].unique()
        if len(teams) != 1:
            abnormal_products.append(i)

    if abnormal_products:
        return "Abnormal products"
    else:
        return "No abnormal products"

result = check_abnormal_products(Df_payment_enriched)
print(result)

```

3.Find the team has had the lowest performance since Q2.2023. Find the category that contributes the least to that team.


```python
def find_weakest_team(df):

    df_q2 = df[df['report_month'] >= '2023-04']

    team_volume = df_q2.groupby('team_own')['volume'].sum()
    weakest_team = team_volume.idxmin()

    team_data = df_q2[df_q2['team_own'] == weakest_team]
    category_volume = team_data.groupby('category')['volume'].sum()
    weakest_cat = category_volume.idxmin()
    cat_data = team_data[team_data['category'] == weakest_cat]

    print("Team has had the lowest performance: " + str(weakest_team))
    print("Category that contributes the least: " + str(weakest_cat))

    return weakest_team, weakest_cat

weakest_team, weakest_cat = find_weakest_team(Df_payment_enriched)
```

4.Find the contribution of source_ids of refund transactions what is the
source_id with the highest contribution?

```python
def find_refund_source_id(df):
    refund_transactions = df[df['payment_group']== 'refund']
    source_volume = refund_transactions.groupby('source_id')['volume'].sum()
    max_source_id = source_volume.idxmax()
    highest_contribution = source_volume.max()
    print("Source_id with the highest contribution: " + str(max_source_id))
    print("Highest contribution: " + str(highest_contribution))
    return max_source_id, highest_contribution
result=find_refund_source_id(payment_report)
```

5.Define type of transactions (‘transaction_type’) for each row, given:
- transType = 2 & merchant_id = 1205: Bank Transfer Transaction
- transType = 2 & merchant_id = 2260: Withdraw Money Transaction
- transType = 2 & merchant_id = 2270: Top Up Money Transaction
- transType = 2 & others merchant_id: Payment Transaction
- transType = 8, merchant_id = 2250: Transfer Money Transaction
- transType = 8 & others merchant_id: Split Bill Transaction
- Remained cases are invalid transactions


```python
def define_transaction_type(row):
    if row['transType'] == 2:
        if row['merchant_id'] == 1205:
            return 'Bank Transfer Transaction'
        elif row['merchant_id'] == 2270:
            return 'Withdraw Money Transaction'
        else:
            return 'Payment Transaction'
    elif row['transType'] == 8:
        if row['merchant_id'] == 2250:
            return 'Transfer Money Transaction'
        else:
            return 'Split Bill Transaction'
    else:
        return 'Invalid Transaction'

Df_transactions['transaction_type'] = Df_transactions.apply(define_transaction_type, axis=1)
Df_transactions.head()


```

6.Of each transaction type (excluding invalid transactions): find the number of transactions, volume, senders and receivers.


```python
def transaction_summary(df):
  df_valid=df[df['transaction_type']!='Invalid Transaction']
  transaction_summary=df_valid.groupby('transaction_type').agg(
      num_transactions=('transaction_type','count'),
      volume=('volume','sum'),
      senders=('sender_id','nunique'),
      receivers=('receiver_id','nunique')
      )
  return transaction_summary
transaction_summary(Df_transactions)
```
